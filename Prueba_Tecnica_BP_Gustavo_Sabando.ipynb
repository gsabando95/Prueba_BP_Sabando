```python
# Ejercicio # 1
cursor = conn.execute('''

WITH PartidoProcesado AS (

  SELECT

    UPPER(TRIM(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(substr(partido, 1, instr(partido, 'vs.') - 1),'ñ','N'), 'á', 'A'), 'í', 'I'), '.', ''), 'ú', 'U'))) AS equipo1,
    UPPER(TRIM(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(substr(partido, instr(partido, 'vs.') + 4),'ñ','N'), 'á', 'A'), 'í', 'I'), '.', ''),'ú', 'U'))) AS equipo2,
    UPPER(grupo) AS grupos,
    CAST(substr(resultado, 1, instr(resultado, ',') -1) AS INTEGER) AS goles1,
    CAST(substr(resultado, instr(resultado, ',') +1) AS INTEGER) AS goles2

  FROM Partidos
),
Puntajes AS(
  SELECT
    equipo1 AS equipo,
    grupos,
    goles1 AS golesFavor,
    goles2 AS golesContra,

    CASE
      WHEN goles1 > goles2 THEN 3
      WHEN goles1 = goles2 THEN 1
      ELSE 0
    END AS puntos
  FROM PartidoProcesado

  UNION ALL

  SELECT
    equipo2 AS equipo,
    grupos,
    goles2 AS golesFavor,
    goles1 AS golesContra,

    CASE
      WHEN goles2 > goles1 THEN 3
      WHEN goles2 = goles1 THEN 1
      ELSE 0
    END AS puntos
  FROM PartidoProcesado
)

SELECT
equipo,
grupos,
SUM(puntos) AS Total_Puntos,
SUM(GolesFavor) - SUM(GolesContra) AS sumatoria_gol_diferencia
FROM Puntajes
GROUP BY
equipo,
grupos
''')


for row in cursor:
  print(row)
```

```python
# Ejercicio # 2
conn = sqlite3.connect('test.db')

####PROGRAME SU RESPUESTA DENTRO DEL CONN.EXECUTE ENTRE LAS COMILLAS

cursor = conn.execute('''

WITH EquiposClasificadosRankeados AS(
  SELECT
    equipo,
    grupo,
    puntos,
    gol_diferencia,

    ROW_NUMBER() OVER(
      PARTITION BY grupo
      ORDER BY puntos DESC, gol_diferencia DESC
    ) AS clasificacionGrupo

    FROM
      Resultados_Qatar


), EquiposClasificados As(
  SELECT
    equipo,
    grupo,
    puntos,
    gol_diferencia,
    clasificacionGrupo

    FROM EquiposClasificadosRankeados
    WHERE clasificacionGrupo <= 2
    ORDER BY grupo, puntos DESC


), GrupoRankingProcesado AS (
  SELECT
    UPPER(TRIM(REPLACE(substr(partido, 1, instr(partido, 'vs.') - 1), 'º',''))) AS grupoRanking1,
    UPPER(TRIM(substr(partido, instr(partido, 'vs.') + 4))) AS grupoRanking2,
    UPPER(TRIM(substr(sede, instr(sede, ',') +1 ))) AS ciudad,
    fecha,
    hora

    FROM Clasificados

), ClasificadosProcesados AS(
  SELECT
    CAST(TRIM(substr(grupoRanking1, 1, instr(grupoRanking1, 'DEL') - 1)) AS INTEGER) AS ranking1,
    UPPER(TRIM(substr(grupoRanking1, instr(grupoRanking1, 'DEL') + 4))) AS grupo1,
    CAST(TRIM(substr(grupoRanking2, 1, instr(grupoRanking2, 'DEL') - 1)) AS INTEGER) AS ranking2,
    UPPER(TRIM(substr(grupoRanking2, instr(grupoRanking2, 'DEL') + 4))) AS grupo2,
    fecha,
    hora,
    ciudad
  FROM GrupoRankingProcesado
), ReporteClasificado AS(

  SELECT
  ec.equipo,
  ec.grupo,
  ec.puntos,
  ec.gol_diferencia,
  cp.fecha,
  cp.hora,
  cp.ciudad
  FROM EquiposClasificados ec
  INNER JOIN ClasificadosProcesados cp
  ON (ec.clasificacionGrupo = cp.ranking1 AND ec.grupo = cp.grupo1) OR (ec.clasificacionGrupo = cp.ranking2 AND ec.grupo = cp.grupo2)

)

  SELECT * FROM ReporteClasificado
                       ''')

for row in cursor:
  print(row)
conn.close()
```

```python
from typing import cast
import sys
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import explode, col, desc, avg, min, max, rank, asc, round
from pyspark.sql import Window
from math import sqrt
from pyspark.sql.functions import udf
from google.colab import files
import json

spark = SparkSession.builder\
        .master("local")\
        .appName("Colab")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()

####### PROGRAME SU RESPUESTA AQUI ##############

!ls /content
ruta = "/content/CENTROS_EDUCATIVOS_MADRID (1).json"
ruta_limpia = "/content/CENTROS_EDUCATIVOS_MADRID_LIMPIA.json"

with open(ruta, "r", encoding="utf-8") as f:
  data = json.load(f)

with open(ruta_limpia, "w",encoding="utf-8") as f:
  for i in data:
    f.write(json.dumps(i)+"\n")

dfCentrosEducativos = spark.read.json(ruta_limpia)
dfCentrosEducativosLimpio = dfCentrosEducativos.where(
    (col("direccion_coor_x").isNotNull()) & (col("direccion_coor_y").isNotNull()))

dfCentrosEducativosLimpio.show()

dfPromedioCoord = dfCentrosEducativosLimpio.groupBy("centro_titularidad").agg(
    avg("direccion_coor_x").alias("prom_coord_x"),
    avg("direccion_coor_y").alias("prom_coord_y")
)
dfPromedioCoord.show()

dfCentrosEducativosPromedio = dfCentrosEducativosLimpio.join(dfPromedioCoord, on='centro_titularidad', how='right')
dfCentrosEducativosPromedio.show()



distanciaUdf = udf(lambda x1, y1, x2, y2: (sqrt(((x2 - x1)**2) + ((y2 - y1)**2))), DoubleType())

dfCentrosEducativosDistancia = dfCentrosEducativosPromedio.withColumn(
    "distancia",
    distanciaUdf(dfCentrosEducativosPromedio["direccion_coor_x"],dfCentrosEducativosPromedio["direccion_coor_y"],dfCentrosEducativosPromedio["prom_coord_x"],dfCentrosEducativosPromedio["prom_coord_y"])
)
dfCentrosEducativosPromedio.show()


ventanaCentroTitularidad = Window.partitionBy("centro_titularidad")

dfCentrosEducativoMenorDist = dfCentrosEducativosDistancia.withColumn("distancia_menor", col("distancia") == min(col("distancia")).over(ventanaCentroTitularidad))
dfCentrosEducativoMenorDist.show()
dfCentrosEducativoMenorDist.select("centro_titularidad","centro_titular","centro_nombre","distancia","distancia_menor").where(col("distancia_menor") == True).show()



#################################################

#EXPORTAR A PARQUET:
dfCentrosEducativoMenorDist.write.mode("overwrite").parquet("/tmp/CENTROS_EDUCATIVOS_MADRID_DISTANCIAS.parquet")

#EXPORTAR A JSON:
dfCentrosEducativoMenorDist.write.mode("overwrite").json("/tmp/CENTROS_EDUCATIVOS_MADRID_DISTANCIAS.json")

#EXPORTAR A CSV:
dfCentrosEducativoMenorDist.write.mode("overwrite").option("header",True).option("sep","|").csv("/tmp/CENTROS_EDUCATIVOS_MADRID_DISTANCIAS.csv")

!ls /tmp/CENTROS_EDUCATIVOS_MADRID_DISTANCIAS.parquet
!ls /tmp/CENTROS_EDUCATIVOS_MADRID_DISTANCIAS.json
!ls /tmp/CENTROS_EDUCATIVOS_MADRID_DISTANCIAS.csv

# 5.3 Justificación:
#Para manejar grandes volumenes de datos y que no requieran ser comprendida su informacion por simple inspeccion humana recomiendo utilizar praquet, posee muchas ventajas en temas de
#almacenamiento y velocidad de acceso a la informacion contenida, ocupa poco espacio en comparacion con los archivos json y csv, adicional puede almacenar tipos de datos variados
#lo cual, no lo puede realizar otros ficheros como los ficheros csv y los ficheros json.
#################################################
```
